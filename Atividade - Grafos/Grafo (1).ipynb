#**Atividade Grafo - Parte 1**


- Implementação das classes de grafo por matriz de adjacência e lista de adjacência.
- Implementação das funções bfs e dfs para os respectivos grafos


```python
# -*- coding: utf-8 -*-
import queue


class Graph:
    def __init__(self, n):
        self.num_vertices = n
        self.matrix = [[0 for i in range(n)] for i in range(n)]
        self.list = [[] for j in range(n)]


    def print_graph(self):
        print(self.matrix)
        print(self.list)


    def bfs(self, source):
        dist = [-1 for _ in range(self.num_vertices)]
        ant = [-1 for _ in range(self.num_vertices)]
        isVisited = [False for _ in range(self.num_vertices)]
        Q = queue.Queue()
        Q.put(source)
        isVisited[source] = True
        dist[source] = 0

        while Q.empty() != True:
            p = Q.get()
            print("Vertice: " + str(p))

            for v in self.list[p]:
                if isVisited[v] == False:
                    dist[v] = dist[p] + 1
                    ant[v] = p
                    Q.put(v)
                    isVisited[v] = True

        return dist, ant

    def dfs(self, inicio):
            visitados = set()
            pilha = [inicio]

            while pilha:
                vertice = pilha.pop()
                if vertice not in visitados:
                    print(vertice, end=' ')
                    visitados.add(vertice)

                    for vizinho in self.list[vertice]:
                        if vizinho not in visitados:
                            pilha.append(vizinho)





    def list_adj(self):
        print("Lista de adjacências")
        for i in range(self.num_vertices):
            print("{} -> {} ".format(i, self.list[i]))


    def le_arq(arquivo): #Lê o arquivo e cria um grafo a partir dele
      arq = open(arquivo, 'r')
      i = int(arq.readline())

      g = Graph(i)

      linha = 0

      for line in arq:
          line = line.strip()
          numeros = line.split("\t")
          coluna = 0
          for i in numeros:
              if(coluna == g.num_vertices):
                  break
              g.matrix[linha][coluna] = int(i)
              if(g.matrix[linha][coluna] > 0):
                  g.list[linha].append(coluna)
              coluna += 1
          linha += 1
      return g


```

#**Grafo - Parte 2**

- Caminhos

```python

def caminho(file, inicio, fim):
    caminho = []
    caminho.append(inicio)

    g = le_arq(file) # cria o grafo

    dist = [-1 for _ in range(g.num_vertices)]
    ant = [-1 for _ in range(g.num_vertices)]
    isVisited = [False for _ in range(g.num_vertices)]
    Q = queue.Queue()
    Q.put(inicio)
    isVisited[inicio] = True
    dist[inicio] = 0 #inicial começa igual a 0

    while Q.empty() != True:
        p = Q.get() # retira

        if p == fim:
            if p not in caminho:
                caminho.append(p)
            print("Caminho: ", caminho)
            return True

        for v in g.list[p]:
            if isVisited[v] == False:
                dist[v] = dist[p] + 1
                ant[v] = p
                Q.put(v)
                isVisited[v] = True

                if fim in g.list[v] or v == fim:
                    if ant[v] not in caminho:
                        caminho.append(ant[v])
    return False
```

#**Grafo - Parte 3 - Teste pcv4**

- Implementação "main"


```python
# Grafo utilizado
g = le_arq("/content/pcv4.txt")
g.print_graph()
g.list_adj()

#Utilizando bfs
dist, ant = g.bfs(1)
print(dist)
print(ant)
caminho(r"/content/pcv4.txt",1,0)
#utilizando dfs
print("\nBusca DFS:")
g.dfs(1)
```
