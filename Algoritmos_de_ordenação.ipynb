#**Selection Sort**
---
- O Algoritmo pega ordena uma lista de elementos colocando os de menor valor a esquerda
- A ordem de complexidade desse algoritmo é de O(n^2) pois temos duas camadas de loops que processam os dados, o mais externo que realiza N ações e o mais interno que diminui de tamanho mas tambem realiza N ações.


```python
def selection_sort(lista):
    n = len(lista)
    for j in range(n-1):
        min_index = j
        for i in range(j, n):
            if lista[i] < lista[min_index]:
                min_index = i
        if lista[j] > lista[min_index]:
            aux = lista[j]
            lista[j] = lista[min_index]
            lista[min_index] = aux

#teste:

'''data = [-2, 45, 0, 11, -9]
size = len(data)
selectionSort(data, size)
print('Vetor em ordem crescente:')
print(data)'''


```

# **Insertion Sort**
---
- O algoritmo funciona comparando valores a um valor inicial e decidindo se os valores comparados sao maiores ou menores que o incial, colocando os menores a esquerda e os maiores a direita.
- Assim como no selection sort a complexidade do algoritmo é de O(n^2) devido as comparações e trocas que são realizadas.



```python
def insertionSort(vet):

    for passo in range(1, len(vet)): #
        key = vet[passo]
        j = passo - 1

        # Compara o valor com cada elemento a sua esquerda ate que um menor seja encontrado
        # Decrescente  key<array[j] vira key>array[j].
        while j >= 0 and key < vet[j]:
            vet[j + 1] = vet[j]
            j = j - 1

        # Adiciona o valor ao lado direito do elemento menor que ele
        vet[j + 1] = key

#teste:
'''data = [9, 5, 1, 4, 3]
insertionSort(data)
print('Vetor em ordem crescente:')
print(data)'''
```

# **Merge Sort**
---
- Baseado no metodo de implementação "divide and conquer", a ideia principal do algoritmo é separar nosso conjunto de dados em "dados menores" e, após a separação, uni-los novamente de forma ordenada.
- Se uma lista tem 8 elementos, com o merge sort vamos separa-la em duas listas de 4 elementos, depois em 4 listas de dois elementos e por fim 8 listas de 1 elemento. Após essa divisão, os itens sao comparados e então ordenados, sempre trabalhando com as sublistas.
- O algoritmo merge sort tende a ser mais rápido que os outros dois trabalhados nesse arquivo, uma vez que sua complexidade é de O(n log n).

```python
def mergesort(lista, inicio=0, fim=None):
    if fim is None:
        fim = len(lista)
    if(fim - inicio > 1): #começo da divisão
        meio = (fim + inicio)//2
        mergesort(lista, inicio, meio) # Algoritmo sendo aplicado na lista esquerda
        mergesort(lista, meio, fim) #Algoritmo sendo aplicado na lista esquerda
        merge(lista, inicio, meio, fim)

def merge(lista, inicio, meio, fim): # Ordenação e junção em uma lista só
    left = lista[inicio:meio] #lista esquerda
    right = lista[meio:fim] #lista direita
    top_left, top_right = 0, 0
    for k in range(inicio, fim):
        if top_left >= len(left): # Se ja tiver chegado ao tamanho maximo da lista esquerda
            lista[k] = right[top_right]
            top_right = top_right + 1
        elif top_right >= len(right): # Se ja tiver chegado ao tamanho maximo da lista direita
            lista[k] = left[top_left]
            top_left = top_left + 1
        elif left[top_left] < right[top_right]: # Se o lado esquerdo tiver um valor menor que o direito
            lista[k] = left[top_left]
            top_left = top_left + 1
        else:
            lista[k] = right[top_right] #se o lado direito tiver um valor menor que o da esquerda
            top_right = top_right + 1
```
